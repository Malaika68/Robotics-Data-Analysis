```python
df['VelocityX']=df['VelocityX']*100
df['VelocityY']=df['VelocityY']*100
# Create a dictionary with the old column names as keys and new column names as values
column_mapping = {'VelocityX': 'Vx', 'VelocityY': 'Vy','BxGain':'Bxx','ByGain':'Byy','BxyGain':'Bxy','ByxGain':'Byx'}
# Use the rename() method to rename the columns
df.rename(columns=column_mapping, inplace=True)
df['Fx']=df['Bxx']*df['Vx'] + df['Bxy']*df['Vy']
df['Fy']=df['Byx']*df['Vx'] + df['Byy']*df['Vy']
# Calculate force using the formula
df['Force'] = np.sqrt(df['Fx']**2 + df['Fy']**2)
df.reset_index(inplace=True)
df.rename(columns={'index': 'RealID'}, inplace=True)
df
df.head()
```

```python
df_0_drone=df_gp_drone[df_gp_drone['Game_name']=='Flying Drone']
#Drone
gp_c=[]
force_mean=[]
count_drone=0
for index, row in df_0_drone.iterrows():
    start_date = row['Start']
    end_date = row['End']
    condition = (df['TimeStampDate'] >= start_date) & (df['TimeStampDate'] <= end_date)
    result_0 = df[condition]
    # Calculate count when x or y changes
    count = result_0[(result_0['VisualTargetXCoord'].diff() != 0) | (result_0['VisualTargetYCoord'].diff() != 0)].shape[0]
    chunk_repeat=(result_0[(result_0['VisualTargetXCoord'].diff() != 0) | (result_0['VisualTargetYCoord'].diff() != 0)])

    mean_list=[];
    for i, id in enumerate(chunk_repeat['RealID']):
        #print(id)
        if i < len(chunk_repeat['RealID'])-1:
            start_index=id
            end_index=chunk_repeat['RealID'].iloc[i+1]-1
            #print(start_index)
            #print(end_index)
            # calculate the mean for the current group
            mean = df.loc[start_index:end_index, 'Force'].mean()
            # add the mean to the list
            mean_list.append(mean)
        elif i == (len(chunk_repeat['RealID'])-1):
            start_index=id
            end_index=result_0['RealID'].iloc[-1]
            #print(start_index)
            #print(end_index)
            # calculate the mean for the current group
            mean = df.loc[start_index:end_index, 'Force'].mean()
            # add the mean to the list
            mean_list.append(mean)
    force_mean.append(mean_list)    
    count_drone=count_drone+count
count_drone
```
